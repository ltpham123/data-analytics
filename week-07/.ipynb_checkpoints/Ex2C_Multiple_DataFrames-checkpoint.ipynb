# Exercise 2.C Working with multiple DataFrames

```python
import numpy as np
import pandas as pd
```

## Basic Aggregations

```python
data = {
    'Product': ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C'],
    'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East',
    'West', 'North'],
    'Units_Sold': [10, 15, 20, 5, 10, 15, 10, 5, 15],
    'Price': [100, 200, 150, 100, 200, 150, 100, 200, 150]
}
df_sales = pd.DataFrame(data)
```

```python
df_sales["Revenue"] = df_sales["Units_Sold"] * df_sales["Price"]
```

```python
print(f"Total units sold: {df_sales["Units_Sold"].sum()}")
print(f"Total revenue: {df_sales["Revenue"].sum()}")
```

```python
print(f"Average units sold: {df_sales["Units_Sold"].mean()}")
print(f"Average revenue: {df_sales["Revenue"].mean()}")
```

## Grouping Data

```python
print(df_sales.groupby("Product")["Units_Sold"].sum())
print(df_sales.groupby("Product")["Revenue"].sum())
```

```python
print(df_sales.groupby("Region")["Units_Sold"].mean())
print(df_sales.groupby("Region")["Revenue"].mean())
```

```python
print(df_sales.groupby(["Product", "Region"])["Units_Sold"].sum())
print(df_sales.groupby(["Product", "Region"])["Revenue"].sum())
```

## Sorting and Ranking Data

```python
df_sales.sort_values(by="Units_Sold", ascending=True)
```

```python
df_sales.sort_values(by="Revenue", ascending=False)
```

```python
df_sales["Revenue_Rank"] = df_sales["Revenue"].rank(ascending=False)
```

```python
df_sales
```

## Merging DataFrames

```python
product_data = {
    'Product': ['A', 'B', 'C'],
    'Category': ['Electronics', 'Furniture', 'Appliances']
}

df_products = pd.DataFrame(product_data)
```

```python
df_merged = pd.merge(df_sales, df_products, on="Product")
```

```python
df_merged
```

```python
print(df_merged.groupby("Category")["Revenue"].sum())
```

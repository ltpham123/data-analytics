# Exercise 1.A NumPy: Manipulating Arrays

```python
import numpy as np
```

## Looping Through Arrays

```python
arrA = np.array([(11, 12, 13), (14, 15, 16)])
print(arrA)
```

```python
for x in arrA:
    print(x * 2)
```

```python
for x in arrA:
    for y in x:
        print(y * 2)
```

Looping through x affects the entire array and doubles each element.
Looping through y affects each individual element in each array, one by one.

```python
for x in arrA:
    for y in x:
        loop_result = x * 2
        print(loop_result)
```

Instead of printing each element, it doubles and prints the entire array. Beginning with the first array, it saves the doubled version and prints it. Each iteration overrides the previous, and since the original is not changed, it continously prints the array of which it belongs to (x).

## Joining and Stacking Arrays

```python
joined_array_1 = np.concatenate((arrA, arrA))
print(joined_array_1)
```

```python
joined_array_2 = np.concatenate((arrA, arrA), axis=1)
print(joined_array_2)
```

```python
stacked_array_1 = np.stack((joined_array_1, joined_array_1))
print(stacked_array_1)
```

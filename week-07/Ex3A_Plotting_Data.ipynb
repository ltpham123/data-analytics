# Exercise 3.A Getting familiar with Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Creating Basic Plots

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, marker='o')
plt.show()
```

```python
plt.plot(x, y, marker='o')
plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

## Styling Plots

```python
plt.plot(x, y, marker='o', linestyle='--', color='red')
plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
plt.plot(x, y, marker='s', markersize=10, linestyle='--', color='green')
plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
plt.plot(x, y, marker='.', markersize=10, linestyle='--', color='blue')
plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
plt.plot(x, y, marker='D', markersize=10, linestyle='--', color='c')
plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

# Exercise 3.B More plotting practice

## Advanced Plotting Techniques

```python
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [23, 17, 35, 29]
```

```python
plt.bar(products, sales)
plt.show()
```

```python
plt.bar(products, sales, width=0.5, color='skyblue')
plt.title("Product Sales")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y)
plt.show()
```

```python
plt.scatter(x, y, s=np.random.rand(50)*100, c=np.random.rand(50), cmap="viridis")
plt.show()
```

## Subplots

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

axes[0].plot(x, y)
axes[0].set_title("Sine Wave Line Chart")
axes[0].set_xlabel("x")
axes[0].set_ylabel("sin(x)")
```

```python
sample_data = np.random.randn(1000)

axes[1].hist(sample_data, bins=30)
axes[1].set_title("1000 Random Numbers Histogram")
axes[1].set_xlabel("Number")
axes[1].set_ylabel("Frequency")
```

```python
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

x = np.linspace(0, 10, 100)
y = np.sin(x)

axes[0].plot(x, y)
axes[0].set_title("Sine Wave Line Chart")
axes[0].set_xlabel("x")
axes[0].set_ylabel("sin(x)")

sample_data = np.random.randn(1000)

axes[1].hist(sample_data, bins=30)
axes[1].set_title("1000 Random Numbers Histogram")
axes[1].set_xlabel("Number")
axes[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

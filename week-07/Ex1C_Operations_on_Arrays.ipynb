## Sample Arrays

```python
r_1 = np.random.randint(1, 21, size=2)
r_2 = np.random.randint(1, 21, size=(1, 2))
r_3 = np.random.uniform(0, 1, size=(2, 1))
r_4 = np.random.randint(20, 101, size=(4, 5))
r_5 = np.random.uniform(0, 1, size=(4, 5))
```

## Permutations

```python
print(r_4)
```

```python
np.random.shuffle(r_4)
print(r_4)
```

Each row was kept the same; however, the order in which they appeared changed.

```python
new_shuffle = np.random.permutation(r_4)
print(new_shuffle)
print(r_4)
```

'np.random.shuffle()' randomizes and modifies the original array permanently
'np.random.permutation()' randomizes the original array too, but unlike .shuffle(), it does not modify the original array permanently.

## Mathematical Operations

```python
print(r_1)
print(r_1.shape)

print(r_2)
print(r_2.shape)

print(np.add(r_1, r_2))
print(np.add(r_1, r_2).shape)
```

```python
print(np.add(np.array([1]), r_4))  # adds 1 to all elements in each array
```

```python
print(np.add(np.array([[1]]), r_4))  # adds 1 to all elements in each array, dimensions are 1, so no conflict happens
```

```python
print(np.add(np.array([1,1,1,1,1]), r_4))  # columns match up, adds 1 to all rows for each corresponding column
```

```python
#print(np.add(np.array([[1], [1]]), r_4))  # columns don't match up, 2x1 vs 4x5
```

NumPy reads the shape of arrays from right to left. If they're the same or one is equal to 1, NumPy extends the shape, or broadcasts, the smaller array to match the larger one. The add function depends on how the arrays line up after broadcasting, and they are arithmetically added based on their corresponding position. The shape of the final array is determined by the largest array of the two.

## Rounding

```python
print(r_3)
```

```python
print(np.trunc(r_3))  # cuts the decimal portion off
```

```python
print(np.fix(r_3))  # rounds towards zero, works for both positive and negative numbers
```

```python
print(np.floor(r_3))  # rounds down to the nearest integer
```

```python
print(np.ceil(r_3))  # rounds up to the nearest integer
```

```python
print(np.round(r_3, 2))  # rounds to 2 decimal places
```

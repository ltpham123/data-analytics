```python
import numpy as np
import pandas as pd
```

## Series

```python
aa = [1, 2, 3, 4]
ab = (5, 6, 7, 8)
ac = {9, 10, 11, 12}
ad = {'first': 13, 'second': 14, 'third': 15, 'fourth': 16}
```

```python
ser1 = pd.Series(aa)
ser2 = pd.Series(ab)
ser3 = pd.Series(list(ac))  # included list() since the order may change any time; TypeError: 'set' type is unordered, a set is always unordered when series requires a fixed order
ser4 = pd.Series(ad)

print(ser1)
print(ser2)
print(ser3)
print(ser4)
```

```python
# ser5 = pd.Series(17, 18, 19, 20)  #TypeError: Index(...) must be called with a collection of some kind, 18 was passed; the argument needs to be a collection, like a list, tuple, or dictionary, before being passed through
```

## DataFrames from Series

```python
df1 = pd.DataFrame(ser1)
df1
```

```python
df1.insert(1, "ser2", ser2)
df1.insert(2, "ser3", ser3)
df1
```

```python
df1.insert(3, "ser4", ser4)
df1  # when ser4 was added, it automatically assumed that ser4 was going to be numbers too. Since this is not the case, it shows up as NaN; pandas matches values by indexes, which means that it couldn't find the value (0 != first)
```

```python
df1.loc[4] = [5, 9, 12, 0]
df1
```

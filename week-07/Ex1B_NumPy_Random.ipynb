# Exercise 1.B NumPy Random

```python
import numpy as np
```

## Generating Random Numbers

```python
r_1 = np.random.randint(1, 21, size=2)
print(r_1)
```

```python
r_2 = np.random.randint(1, 21, size=(1, 2))
print(r_2)
```

```python
r_3 = np.random.uniform(0, 1, size=(2, 1))
print(r_3)
```

```python
r_4 = np.random.randint(20, 101, size=(4, 5))
print(r_4)
```

```python
r_5 = np.random.uniform(0, 1, size=(4, 5))
print(r_5)
```

## Generating Random Choices from an Array

```python
students = np.array([
    "Onur", "Jessica", "Julissa", "Jimmy", "Maritza",
    "Tracy", "Paul", "Kiki", "Will", "Shaq",
    "Fuat", "El" # , "Rae", "Angel", "Ceyhun",
    #"Shay", "Paola", "Eyerusalem", "Cagla", "David"
])
```

```python
call_on = np.random.choice(students)
print(call_on)
```

```python
breakout_groups = np.random.choice(students, size=(3, 4))
print(breakout_groups)  # not quite an effective way to assign breakout rooms because some rooms have the same participants
```

```python
p_values = [0.10, 0.30, 0.08, 0.00, 0.10, 0.05, 0.08, 0.05, 0.05, 0.05, 0.05, 0.09]  # based on how likely I am to collaborate with them on a future project

call_on_weighted = np.random.choice(students, size=5, replace = False, p = p_values)  # added replace condition to pick without replacement

print(call_on_weighted)
```

## Generating a Normal Distribution

```python
bell_curve = np.random.normal(loc = 10, scale = 5, size = (4, 5))
print(bell_curve)
```

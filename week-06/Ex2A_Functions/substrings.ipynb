```python
def parse_and_display(name):
    for i in range(len(name)):
        if name[i] == ' ':
            first = name[:i]
            last = name[i + 1:]
            print(f'Full name: {first} {last}')
            print(f'First name: {first}')
            print(f'Last name: {last}')
```

```python
parse_and_display('Barnacle Bill')
```

```python
parse_and_display('John Doe')
```

```python
def parse_and_display2(name):
    full_name = name
    first = name  # assigns the name automatically as first name
    middle = ''
    last = ''
    for i in range(len(name)):
        if name[i] == ' ':  # detects first whitespace
            if last == '':
                first = name[:i]
                last = name[i + 1:]
                full_name = f'{first}' + f' {last}'
            else:            
                middle = name[len(first) + 1:i]
                last = name[i + 1:]
                full_name = f'{first}' + f' {middle}' + f' {last}'
    print(f'Full name: {full_name}')
    print(f'First name: {first}')
    if middle != '':
        print(f'Middle name: {middle}')
    if last != '':
        print(f'Last name: {last}')
```

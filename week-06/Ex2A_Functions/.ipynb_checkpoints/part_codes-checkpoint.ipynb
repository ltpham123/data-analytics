# Exercise 2.A Using built-in functions
## Lab 2. Parsing strings - part codes

```python
def get_supplier_code(partcode):  # before :
    for i in range(len(partcode)):
        if partcode[i] == ":":
            return partcode[:i]
```

```python
def get_product_num(partcode):  # between : and -
    for i in range(len(partcode)):
        if partcode[i] == ":":
            colon_index = i
        if partcode[i] == "-":
            return partcode[colon_index + 1:i]
```

```python
def get_size(partcode):  # after -
    for i in range(len(partcode)):
        if partcode[i] == "-":
            return partcode[i + 1:]
```

```python
test_partcode = ["ACME:123-L", "DI:12345-M", "ACE:1-12"]
```

```python
print(get_supplier_code(test_partcode[0]))
print(get_supplier_code(test_partcode[1]))
print(get_supplier_code(test_partcode[2]))
```

```python
print(get_product_num(test_partcode[0]))
print(get_product_num(test_partcode[1]))
print(get_product_num(test_partcode[2]))
```

```python
print(get_size(test_partcode[0]))
print(get_size(test_partcode[1]))
print(get_size(test_partcode[2]))
```

```python
print(f"For {test_partcode[0]}, the supplier is {get_supplier_code(test_partcode[0])}. Part number {get_product_num(test_partcode[0])} is available in size {get_size(test_partcode[0])}.")
```

```python
print(f"For {test_partcode[1]}, the supplier is {get_supplier_code(test_partcode[1])}. Part number {get_product_num(test_partcode[1])} is available in size {get_size(test_partcode[1])}.")
```

```python
print(f"For {test_partcode[2]}, the supplier is {get_supplier_code(test_partcode[2])}. Part number {get_product_num(test_partcode[2])} is available in size {get_size(test_partcode[2])}.")
```

```python
test_partcode_list = []
test_partcode_list.append([get_supplier_code(test_partcode[0]), get_product_num(test_partcode[0]), get_size(test_partcode[0])])
test_partcode_list.append([get_supplier_code(test_partcode[1]), get_product_num(test_partcode[1]), get_size(test_partcode[1])])
test_partcode_list.append([get_supplier_code(test_partcode[2]), get_product_num(test_partcode[2]), get_size(test_partcode[2])])
test_partcode_list
```

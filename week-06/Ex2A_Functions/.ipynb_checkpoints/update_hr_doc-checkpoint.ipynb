## Lab 3. Employee data
This lab exercise makes updates to information extracted from an employee database

```python
hr_list = [
    ('0123', 'HR', 'Rebecca Yang', '69000'),
    ('0121', 'IT', 'Mark Blick', '67000'),
    ('0068', 'IT', 'Kahna Larsen', '112000'),
    ('0234', 'OPS', 'Jim Smith', '54000'),
]
```

```python
hr_list[1] = hr_list[1][:2] + ('Mark Blick-Hawley',) + hr_list[1][3:]
```

```python
print(hr_list)
```

```python
hr_list[3] = (hr_list[3][0],) + ('CS',) + (hr_list[3][2],) + ('60000',)
```

```python
print(hr_list)
```

```python
for employee in hr_list:
    id_num = employee[0]
    dept_code = employee[1]
    emp_name = employee[2]

    salary = format(int(employee[3]), ",")

    print(f"{id_num} | {dept_code} | {emp_name} | {salary}")
```

# Exercise 5: NumPy Fundamentals
This exercise demonstrates how to import and use NumPy for data analytics related purposes

```python
import numpy as np
```

## Array Basics

```python
# step a
fruits = ["apple", "banana", "cherry", "date"]
```

```python
# step b
arr_1 = np.array(fruits)
```

```python
# step c
arr_1.dtype  # returns '<U6'; U is unicode string data type, 6 means longest string element in array is 6 characters
```

```python
# step d
arr_2 = np.array([(10, 11, 12), (13, 14, 15)])
print(arr_2)
```

```python
# step e
print(arr_2.T)  # transposes arr_2
```

```python
print(arr_2)  # no change made after transposing
```

```python
# step f
arr_3 = np.array([(1, 3, 5), (2, 4, 6)])
print(arr_3)
```

```python
# step g
arr_4 = np.array([arr_3[0, 2], arr_3[1, 1]])
print(arr_4)
```

```python
# step h
arr_5 = np.array([[(1, 1), (1, 1)], [(1, 1), (1, 1)]])

# or
# arr_5 = np.ones((2, 2, 2), dtype=int)  # 2 x 2 x 2; 3D array with ones
print(arr_5)
```

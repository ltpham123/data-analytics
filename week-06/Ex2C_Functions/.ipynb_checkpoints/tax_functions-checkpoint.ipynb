# Exercise 2C: Functions
## Lab 3: Exploring more custom functions into taxes

```python
def get_soc_sec_tax(gross_pay):
    return gross_pay * .062
```

```python
def get_medicare_tax(gross_pay):
    return gross_pay * .0145
```

```python
def get_federal_tax(gross_pay, withholding_code):
    wh_percent = 0
    if withholding_code == 0:
        wh_percent = .23
        return gross_pay * wh_percent
    elif withholding_code == 1:
        wh_percent = .21
        return gross_pay * wh_percent
    elif withholding_code == 2:
        wh_percent = .195
        return gross_pay * wh_percent
    elif withholding_code == 3:
        wh_percent = .185
        return gross_pay * wh_percent
    elif isinstance(withholding_code, int) and withholding_code >= 4:
        wh_percent = .18 - ((withholding_code - 4) * .005)
        return gross_pay * wh_percent
    else:
        print("Invalid withholding code. Please try again.")
```

```python
person1 = {
    "gross_pay": 750,
    "withholding_code": 0,
}

person2 = {
    "gross_pay": 1550,
    "withholding_code": 2,
}

person3 = {
    "gross_pay": 1100,
    "withholding_code": 5,
}
```

```python
# person 1
print(f"Person 1 - Tax taken: ${get_soc_sec_tax(person1['gross_pay']) + get_medicare_tax(person1['gross_pay']) + get_federal_tax(person1['gross_pay'], person1['withholding_code']):.2f}")

# person 2
print(f"Person 2 - Tax taken: ${get_soc_sec_tax(person2['gross_pay']) + get_medicare_tax(person2['gross_pay']) + get_federal_tax(person2['gross_pay'], person2['withholding_code']):.2f}")

# person 3
print(f"Person 3 - Tax taken: ${get_soc_sec_tax(person3['gross_pay']) + get_medicare_tax(person3['gross_pay']) + get_federal_tax(person3['gross_pay'], person3['withholding_code']):.2f}")
```

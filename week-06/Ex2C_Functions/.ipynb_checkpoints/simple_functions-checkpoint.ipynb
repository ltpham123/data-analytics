## Lab 1: Defining Custom Functions

```python
def favorite_things(name, movie, music):
    print(f"Hi {name}!")
    print(f"Your favorite movie is {movie}.")
    print(f"Your favorite music is {music}.")
```

```python
favorite_things("Longley", "The Imitation Game", "YOASOBI")
```

```python
def why_im_here(purpose):
    print(f"Why I joined the Year Up United Data Analyst program: {purpose}")
```

```python
why_im_here("To enter the corporate scene, secure a job, network, and provide for my family.")
```

```python
def favorite_place(place, why):
    print(f"I love to visit {place} because {why}")
```

```python
favorite_place("Vietnam", "I love to go back to my culture, eat food, explore new places and meet new faces, and revisit family.")
```

```python
favorite_place("Japan", "I'm interested in Japanese culture and food, and how its become tourist-like with lots of sightseeing to do!")
```

### Exercise 2.E Using functions for analyzing data
This notebook analyzes a small sales dataset using Python functions.
Each function takes the full dataset as its first argument.

```python
sales_data = [
    ('Laptop', 'East', 1200.00),
    ('Phone', 'West', 850.00),
    ('Tablet', 'East', 400.00),
    ('Monitor', 'South', 650.00),
    ('Keyboard', 'West', 75.00),
    ('Laptop', 'South', 1150.00),
    ('Phone', 'East', 900.00),
    ('Tablet', 'West', 425.00),
    ('Monitor', 'East', 680.00),
    ('Keyboard', 'South', 80.00),
    ('Laptop', 'West', 1300.00),
    ('Phone', 'West', 820.00),
]
```

## Lab 1. Total sales by region

```python
def region_total(data, region):
    total = 0
    for sales in data:
        if sales[1] == region:
            total += sales[2]
    return total
```

```python
print(region_total(sales_data, 'East'))
print(region_total(sales_data, 'West'))
print(region_total(sales_data, 'South'))
```

## Lab 2. Count of sales by region

```python
def region_count(data, region):
    total = 0
    for sales in data:
        if sales[1] == region:
            total += 1
    return total
```

```python
print(region_count(sales_data, 'East'))
print(region_count(sales_data, 'West'))
print(region_count(sales_data, 'South'))
```

## Lab 3. Average sale amount by region

```python
def region_average(data, region):
    if region_count(data, region):
        return region_total(data, region) / region_count(data, region)
    else:
        return 0
```

```python
print(f"{region_average(sales_data, 'East'):.2f}")  # East
print(f"{region_average(sales_data, 'West'):.2f}")  # West
print(f"{region_average(sales_data, 'South'):.2f}")  # South
print(f"{region_average(sales_data, 'North'):.2f}")  # North
```

## Lab 4. Filter by sale amount

```python
def sales_above(data, threshold):
    filtered_sales = []
    for sale in data:
        if sale[2] > threshold:
            filtered_sales.append(sale)
    return filtered_sales
```

```python
for sale in sales_above(sales_data, 500):
    print(sale)
```

```python
def sales_below(data, threshold):
    filtered_sales = []
    for sale in data:
        if sale[2] < threshold:
            filtered_sales.append(sale)
    return filtered_sales
```

```python
for sale in sales_below(sales_data, 500):
    print(sale)
```

## Lab 5. Top and bottom sale

```python
# def top_sale(data):
#     highest_sale = data[0]
#     for sale in data:
#         if sale[2] > highest_sale[2]:
#             highest_sale = sale
#     return highest_sale

# def bottom_sale(data):
#     lowest_sale = data[0]
#     for sale in data:
#         if sale[2] < lowest_sale[2]:
#             lowest_sale = sale
#     return lowest_sale

def top_sale(data):
    return max(data, key=lambda s: s[2])

def bottom_sale(data):
    return min(data, key=lambda s: s[2])
```

```python
print(f"The top sale was a {top_sale(sales_data)[0]} in the {top_sale(sales_data)[1]} region for ${top_sale(sales_data)[2]:.2f}.")
print(f"The bottom sale was a {bottom_sale(sales_data)[0]} in the {bottom_sale(sales_data)[1]} region for ${bottom_sale(sales_data)[2]:.2f}.")
```

## Lab 6. Apply a discount

```python
def apply_discount(data, discount_pct):
    return list(map(lambda s: (s[0], s[1], s[2] * (1 - discount_pct)), data))
```

```python
print(apply_discount(sales_data, .15))
```

## Lab 7. Unique regions

```python
def unique_regions(data):
    regions = [data[0][1]]
    for sale in data:
        if sale[1] not in regions:
            regions.append(sale[1])
    return sorted(regions)
```

```python
print(unique_regions(sales_data))
```

Since this is a unique regions function, I can use this to describe the sales data's total coverage. We see that there is no sales in the North region, so we might infer and predict to make sales there to have more, if not full coverage.

## Lab 8. Region summary

```python
def region_summary(data, region):
    region_total(data, region)
    region_count(data, region)
    region_average(data, region)

    region_data = [sale for sale in data if sale[1] == region]
    top_sale(region_data)
    print(f"=== {region.capitalize} Region ===")
    print(f"Records:\t{region_count(data, region)}")
    print(f"Total Sales:\t{region_total(data, region):.2f}")
    print(f"Average Sales:\t{region_average(data, region):.2f}")
    print(f"Top Sale:\t{top_sale(region_data)[0]}—{top_sale(region_data)[2]:.2f}")
```

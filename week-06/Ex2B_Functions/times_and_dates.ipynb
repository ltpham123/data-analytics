## Lab 2. Datetime Functions
This lab exercise works through with date and time functions grabbed from the datetime library.

```python
import datetime
```

# Today

```python
# current date and time
today = datetime.datetime.now()
```

```python
print(f"Today is {today.strftime("%A, %B %d, %Y")}")
```

# My birthday

```python
# my birthday: Jul 13 2000
birthdate = datetime.date(2000, 7, 13)
print(f"My birthday is {birthdate.strftime("%m/%d/%Y")}")
```

# Next 90 Days

```python
ninety_d = today + datetime.timedelta(days=90)
```

```python
print(f"90 days from today is {ninety_d.strftime("%B %d, %Y")}")
```

# Time for Dinner

```python
dinner_time = datetime.time(19, 0, 0)
```

```python
print(f"Let's meet for dinner at {dinner_time.strftime("%I:%S %p")}")
```
